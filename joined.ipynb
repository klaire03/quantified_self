```python
import pandas as pd
from tabulate import tabulate

# Load the dataset
file_path = '/content/joined.csv'
df = pd.read_csv(file_path)

# Display the head of the DataFrame in a table
print("Head of the DataFrame:")
print(tabulate(df.head(), headers='keys', tablefmt='fancy_grid'))
```

```python
# Calculate summary statistics and display in a table
statistics_table = df.describe().transpose()
print("Summary Statistics:")
print(tabulate(statistics_table, headers='keys', tablefmt='fancy_grid'))
```

```python
# Calculate percentage of missing values for each column
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Create a DataFrame to display the results
missing_info = pd.DataFrame({
    'Column': df.columns,
    'Missing Percentage': missing_percentage
})

# Sort the DataFrame by missing percentage in descending order
missing_info = missing_info.sort_values(by='Missing Percentage', ascending=False)

# Display the missing values information
print("Missing Values Summary:")
print(missing_info)
```

```python
# Assuming you have a column named 'content' in your DataFrame
df['content'] = df['content'].fillna(0)

# Now, all NULL values in the 'content' column are replaced with 0
```

```python

# Display the head of the DataFrame in a table
print("Head of the DataFrame:")
print(tabulate(df.head(), headers='keys', tablefmt='fancy_grid'))
```

```python
df = pd.read_csv('joined.csv')  # Replace 'joined.csv' with your actual file path

# Define the moon phase categories and corresponding ranges
categories = ['New Moon', 'First Quarter', 'Gibbous Waxing', 'Full Moon', 'Gibbous Waning', 'Last Quarter', 'New Moon']
ranges = [0.0, 0.1, 0.3, 0.49, 0.51, 0.8, 0.98]

# Replace the 'moonphase' column with integer bins (0 or 1)
df['moonphase'] = pd.cut(df['moonphase'], bins=ranges, labels=categories[:-1], right=False)
df['moonphase'] = (df['moonphase'] == 'Full Moon').astype(int)

# Display the resulting DataFrame
print(df[['moonphase']])

# Save the resulting DataFrame to a new CSV file
df.to_csv('joined.csv', index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Assuming you have a DataFrame named 'df' with your dataset in 'joined.csv'
df = pd.read_csv('joined.csv')  # Replace 'joined.csv' with your actual file path

# Extract numerical columns for normalization
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numerical columns
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Save the normalized DataFrame to a new CSV file
df.to_csv('normalized_joined.csv', index=False)
```

```python

# Display the head of the DataFrame in a table
print("Head of the DataFrame:")
print(tabulate(df.head(), headers='keys', tablefmt='fancy_grid'))
```
